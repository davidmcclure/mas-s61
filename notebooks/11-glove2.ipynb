```python
import numpy as np
import pandas as pd

from scipy.spatial.distance import cosine
from gensim.models import KeyedVectors
from tqdm import tqdm_notebook
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
pd.options.display.max_rows = 1000
```

```python
m1 = KeyedVectors.load_word2vec_format('../data/models/v2/glove/new-england.txt')
```

```python
m2 = KeyedVectors.load_word2vec_format('../data/models/v2/glove/deep-south.txt')
```

```python
data = []
for token in tqdm_notebook(m1.wv.vocab.keys()):
    
    if token not in m2:
        continue
        
    sim1 = [t for t, _ in m1.most_similar(token, topn=50)]
    sim2 = [t for t, _ in m2.most_similar(token, topn=50) if t in m1]
    
    if sim2 < sim1:
        continue
        
    score = m1.n_similarity(sim1, sim2)
    
    data.append((token, score))
```

```python
df = pd.DataFrame(data, columns=('token', 'score'))
```

```python
def compare(token):
    for t, _ in m1.most_similar(token, topn=20):
        print(t)
    print('\n')
    for t, _ in m2.most_similar(token, topn=20):
        print(t)
```
