```python
m1 = KeyedVectors.load_word2vec_format('../data/corpora/north-south/new-england.w2v.txt')
```

```python
m2 = KeyedVectors.load_word2vec_format('../data/corpora/north-south/deep-south.w2v.txt')
```

```python
vocab = parse_glove_vocab('../data/corpora/north-south/vocab.txt')
```

```python
def sim_ratios(m1, m2, seed):
    data = []
    for token, count in vocab.items():
        if count > 50000:
            m1s = m1.similarity(seed, token)
            m2s = m2.similarity(seed, token)
            data.append((token, m1s-m2s))
    return pd.DataFrame(data, columns=('token', 'ratio'))
```

```python
god = sim_ratios(m1, m2, 'life')
```

```python
god.sort_values('ratio', ascending=False)
```
