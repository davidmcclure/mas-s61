```python
import numpy as np
import pandas as pd

from gensim.models import Word2Vec
from scipy.spatial.distance import cosine
from tqdm import tqdm_notebook
from itertools import combinations
from scipy.special import comb
```

```python
pd.options.display.max_rows = 1000
```

```python
ne = Word2Vec.load('../data/northeast.bin')
```

```python
mw = Word2Vec.load('../data/midwest.bin')
```

```python
ne_vocab = set([k for k, v in ne.wv.vocab.items() if v.count > 500000])
```

```python
mw_vocab = set([k for k, v in mw.wv.vocab.items() if v.count > 500000])
```

```python
vocab = set.intersection(ne_vocab, mw_vocab)
```

```python
vocab = sorted(vocab)
```

```python
len(vocab)
```

```python
pairs = combinations(vocab, 2)
total = comb(len(vocab), 2)

data = []
for t1, t2 in tqdm_notebook(pairs, total=total):
    nes = ne.similarity(t1, t2)
    mws = mw.similarity(t1, t2)
    data.append((t1, t2, nes, mws, nes-mws))
```

```python
df = pd.DataFrame(data, columns=('t1', 't2', 'nes', 'mws', 'd'))
```
